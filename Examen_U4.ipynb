### Parte A

#### Diferencia hacia Adelante

```python
import numpy as np # importa la bibloteca numpy para calculos numericos

def f(x): #define la funcion matematica a analizar
  return x**3-6*x**2+9*x+1 #retorna el valor de la funcion x 3

x = 2 # punto donde se calculara la derivada

h = 0.001 #tamaño del paso numerico

derivada = (f(x+h)-f(x))/h # aplica la formula de diferencia hacia adelante

real = 3*x**2-12*x+9 #calcula la derivada analitica real

error = abs(real - derivada) # calcuña el error absoluto

print("Derivada numerica:" ,derivada) # muestra el valor aproximado

print("Derivada real : ",real) # muestra el valor exacto

print("Error : ",error) # muestra la diferencia entre valores
```

#### Diferencia Central

```python
import numpy as np #importa numpy

def f(x): #define funcion matematica
  return x**3-6*x**2+9*x+1 # retorna 3

x = 2 #punto de evaluacion

h = 0.001 #paso pequeño

derivada = (f(x+h)-f(x-h))/(2*h) #formula de diferencia central

real = 3*x**2-12*x+9 # derivada exacta

error = abs(real - derivada) #  error absoluto

print("Deriada numerica:" ,derivada) # resultado numerico

print("derivada real : ",real) # resultado numerico

print("Error : ",error) # muestra la diferencia entre valores
```

### Parte B

#### Método del Trapecio

```python
import numpy as np #importa la bibloteca numpy para calculos
def f(x): # define la funcion a integrar
  return x**3 - 6*x**2 + 9*x + 1 # retorna el valor de la funcion x ** 2

a = 0 # limite inferior de integracion
b = 3 # limite superior de integracion
n = 100 # numero de subintervalo

h = (b-a)/n # calcula el tamaño del paso

x = np.linspace(a,b,n+1) # Genera los puntos de discretizacion del intervalo

y = f(x) # evalua la funcion en todos los puntos

integral = (h/2)*(y[0]+2*sum(y[1:n])+y[n]) #

print("integral aproximada (Trapecio) : ",integral) # Muestra el resutado numerico
```

#### Método de Simpson

```python
import numpy as np #importa la bibloteca numpy para calculos
def f(x): # define la funcion a integrar
  return x**3 - 6*x**2 + 9*x + 1 # retorna el valor de la funcion x ** 2

a = 0 # limite inferior de integracion
b = 3 # limite superior de integracion
n = 100 # numero de subintervalo

h = (b-a)/n # calcula el tamaño del paso

x = np.linspace(a,b,n+1) # Genera los nodos del intervalo

y = f(x) # evalua la funcion

suma = y[0] + y[n] # inicializada la suma con extremos

for i in range(1,n): # itera sobre los puntos interiores
  if i%2 == 0: # verifica si el indice es par
    suma += 2*y[i] # Aplica coeficiente 2
  else:  # si no es par
    suma += 4*y[i] # Aplica coeficiente 4

integral = (h/3)*suma # calcula la integral

print("integral aproximada (Simpson) : ",integral) #Muestra resultado
```

### Parte C

```python
import numpy as np

# Definición de la función de densidad conjunta
def f(x, y):
    return np.exp(-(x**2 + y**2))

#Región y discretización
lim_inf, lim_sup = -1, 1
n_puntos = 1000

x_coords = np.linspace(lim_inf, lim_sup, n_puntos)
y_coords = np.linspace(lim_inf, lim_sup, n_puntos)

# Calcular el diferencial de área dA (dx * dy)
dx = (lim_sup - lim_inf) / (n_puntos - 1)
dy = (lim_sup - lim_inf) / (n_puntos - 1)
dA = dx * dy

X, Y = np.meshgrid(x_coords, y_coords) # Implementar suma doble
Z = f(X, Y) # Creamos coordenadas

integral_doble = np.sum(Z) * dA # La integral es la suma de los valores por el diferencial de área

print(f"Integración Múltiple (Región [ -1, 1 ] x [ -1, 1 ]) ")
print(f"Resultado de la aproximación: {integral_doble:.8f}")
print(f"\nInterpretación:")
print(f"La probabilidad acumulada en esta región específica es de {integral_doble:.4f}")
```
