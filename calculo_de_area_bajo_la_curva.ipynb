# Aplicacion IA : calculo de area bajo la curva

```python
import numpy as np #importa numpy

def f(x): #define la funcion tipo gaussiana
  return np.exp(-x**2) #calcula e^(-x^2)

a = 0 #limite inferior
b = 1 #limite superior
n = 100 #numero de iteraciones

h = (b-a)/n #calcula el tamaño del paso

x = np.linspace(a,b,n+1) #genera puntos

y = f(x) # evalua funcion

integral = (h/2)*(y[0]+2*sum(y[1:n])+y[n]) #Aplica

print("El area bajo la curva : ",integral) #resultado
```
