# Integral doble fisica

```python
import numpy as np #importa numpy
def f(x,y): #define la funcion de dos variables
  return x + y # Retorna la suma de las variables

a,b = 0,1 #Define los limites en el eje x
c,d = 0,1 # define los limites en el eje y

n = 50 # Numero de divisiones en x
m = 50 # Numero de divisiones en y

dx = (b-a)/n # calcula el tamaño del paso en x
dy = (d-c)/m # calcula el tamaño del paso en x

integral = 0  # inicializa el acumulador de la integral

for i in range(n): # itera sobre las participaciones en x
  for j in range(m): # itera sobre las participaciones en y

    x = a + i*dx # calcula el valor de x
    y = c + j*dy # calcula el valor de y

    integral += f(x,y)*dx*dy #Suma el area el elemental

print ("integral doble aproximada : ",integral) #muestra el resultado

```
