# Diferencia central

```python
import numpy as np #importa numpy

def f(x): #define funcion matematica
  return x**3 # retorna 3

x = 2 #punto de evaluacion

h = 0.001 #paso pequeño

derivada = (f(x+h)-f(x-h))/(2*h) #formula de diferencia central

real = 3*x**2 # derivada exacta

error = abs(real - derivada) #  error absoluto

print("Deriada numerica:" ,derivada) # resultado numerico

print("derivada real : ",real) # resultado numerico

print("Error : ",error) # muestra la diferencia entre valores
```
